```python
import numpy as np
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
```

```python
# Compute the softmax
z = [1,2,3]

numerator = np.exp(z)
# denominator = np.sum(np.exp(z))
denominator = np.sum(numerator)
softmax = numerator/denominator
print(softmax)
print(np.sum(softmax))
```

```python
# Logarithm
# is the inverse of the natural exponent 
# exp(x) = log(exp(x))
# log is a monotonic function of x (x up log also up)
# sine wave is non-monotonic function
```

```python
x = np.linspace(0.001,1,200)
# print(x)

logx = np.log(x)

fig = plt.figure(figsize=(10,4))

plt.rcParams.update({'font.size':15})

plt.plot(x,logx,'ks-',markerfacecolor='w')
plt.xlabel('x')
plt.ylabel('log(x)')
plt.show()
```

```python
x = np.linspace(0.0001,1,20)

logx = np.log(x)
expx = np.exp(x)

plt.plot(x,x,color=[.8,.8,.8])
plt.plot(x,np.exp(logx),'o',markersize=8)
plt.plot(x,np.log(expx),'x',markersize=8)
plt.xlabel('X')
plt.ylabel('f(g(x))')
plt.legend(['unity','exp(log(x))','log(exp(x))'])
plt.show()
```
