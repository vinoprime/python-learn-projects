```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib  inline
```

```python
df =  pd.read_csv("../../../data/weight-height.csv")
# df.head()
```

```python
plt.scatter(df["Weight"],df["Height"])
plt.xlabel("Weight")
plt.ylabel("Height")
```

```python
# Correlation
# df.corr(method="pearson", min_periods=1)
import seaborn as sns
sns.pairplot(df)
```

```python
X = df[["Weight"]] ## independent feature should be dataframe or 2d array
type(X)
np.array(X).shape
X.head()

y = df["Height"] ## dependent feature can be series form or 1d array
y
```

```python
# # Train Test split
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.25,random_state=42)
```

```python
X_train.shape
```

```python
# # Standardization
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
X_test
```

```python
## Machine learning algoritham SLR

from sklearn.linear_model import LinearRegression

reg = LinearRegression(n_jobs=-1)
reg.fit(X_train,y_train)

# reg.predict()
reg.coef_
```

```python
print("Coefficient or Slope B1", reg.coef_)
print("Intercept B0", reg.intercept_)
```

```python
## Plot Training data plot best fit line
plt.scatter(X_train,y_train)
plt.plot(X_train, reg.predict(X_train))
```
