```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import Lasso
```

```python
# Assign the data to predictor and outcome variables
# TODO: Load the data
train_data = pd.read_csv('./data/regularization.csv', header = None)
X = train_data.iloc[:,:-1]
y = train_data.iloc[:,-1]

# TODO: Create the linear regression model with lasso regularization.
lasso_reg = Lasso()

# TODO: Fit the model.
lasso_reg.fit(X,y)

# TODO: Retrieve and print out the coefficients from the regression model.
reg_coef = lasso_reg.coef_
# print(reg_coef)


# Create a list of feature names (you may need to replace these with your actual feature names)
feature_names = X.columns

# Retrieve the coefficients from the Lasso regression model
reg_coef = lasso_reg.coef_

# Create a bar plot to visualize the coefficients
plt.figure(figsize=(10, 6))
plt.bar(range(len(feature_names)), reg_coef, tick_label=feature_names)
plt.xlabel("Features")
plt.ylabel("Coefficients")
plt.title("Lasso Regression Coefficients")
plt.xticks(rotation=45, ha="right")  # Rotate feature names for readability
plt.show()
```
